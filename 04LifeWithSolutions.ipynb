## Conway's Game of Life

This is a 2D grid of cells, which can be on or off, 'living' or 'dead'. This grid changes over a series of generations.

Each cell has 8 neighbours (up, down, left, right and the diagonals). The number of neighbours of each cell determines whether it will live in the next generation:

* Cells with less than two neighbours will die
* Cells with two neighbours will stay alive if already alive
* Cells with three neighbours will come to life
* Cells with four or more neighbours will die

### Note about computing bounds

When computing (for example) where the left-hand side of the neighbours of our cell is, we might use something like:

```python
if x-1 < 0:
    left = 0
else:
    left = x-1
```

This is exactly equivalent to:

```python
left = min(0, x-1)
```

If `x-1` is less than zero, then the minimum returned will be zero. Otherwise, it will be `x-1`.

````python
### Update the grid

def update_cell(x, y, grid):
    ## Here are two ways to sum up the 3 by 3 grid around x,y.
    # * Using NumPy indexing. First work out the bounds for our 3 by 3 grid, taking
    # into account the edges:
    left = max(0,x-1)
    right = min(grid.shape[0],x+2)
    top = max(0,y-1)
    bottom = min(grid.shape[1],y+2)
    
    # Now get this 3 by 3 grid out and sum it
    neighbours = grid[left:right,top:bottom]
    s = np.sum(neighbours.flatten())
    
    # * Alternatively, compute a sum for the grid using a loop.
    # (this would replace the two lines of code immediately above)
    #s = 0
    #for x1 in range(left,right):
    #    for y1 in range(top,bottom):
    #        s += grid[x1,y1]
    
    # * In either case, we have included the cell x,y itself in the sum, so we must
    # subtract its value
    s -= grid[x,y]
    
    if s < 2:
        return 0
    elif s == 2:
        return grid[x,y]
    elif s == 3:
        return 1
    else:
        return 0
    
def update_grid(grid):
    new_grid = np.zeros(grid.shape)
    for x in range(grid.shape[0]):
        for y in range(grid.shape[1]):
            new_grid[x,y] = update_cell(x,y,grid)
    return new_grid
````
